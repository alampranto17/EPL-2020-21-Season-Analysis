### Player base on Difference age Group 

```python
def ageOrder(age):
    if age<20:
        return 'under20'
    elif age>=20 and age<25:
        return '20between25'
    elif age>=20 and age<30:
        return '25between30'
    else:
        return 'above30'
    
    
    
        
    
df['Age_level']=df['Age'].apply(ageOrder)
```

```python
listofage=[]

listofage.append(df[df['Age_level']=='under20'].shape[0])
listofage.append(df[df['Age_level']=='20between25'].shape[0])
listofage.append(df[df['Age_level']=='25between30'].shape[0])
listofage.append(df[df['Age_level']=='above30'].shape[0])


listofage
```

```python
import plotly.express as px
fig = px.pie(values=listofage,
             names=['under 20','20-25','25-30','30+'],
             title='Player Distribution by Age')

fig.show()
```

Conclusion
- Most of the players in the EPL are between 25 and 30 years old.

```python
pivot = df.pivot_table(index=['Club','Name'],
                       values=['Goals', 'Assists'],
                       aggfunc='sum').reset_index()

pivot['Total_Contributions'] = pivot['Goals'] + pivot['Assists']
pivot = pivot.sort_values('Total_Contributions', ascending=False)

top_scorers = pivot.groupby('Club').head(1)
top_scorers
```

```python
px.scatter(
    top_scorers,
    x='Goals',
    y='Assists',
    size='Total_Contributions',   # bubble size
    color='Club',                 # optional: color by club/team
    hover_name='Name',            # shows name when you hover
    text='Name',                  # optional: display names
    size_max=50,
    color_discrete_sequence=px.colors.qualitative.Safe
)
```
